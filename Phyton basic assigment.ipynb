 Q1:-  What is Python, and why is it popular?

 Ans:- In short, Python is a versatile and beginner-friendly programming language used for web development, data science, scripting, and more. Its popularity stems from its clear syntax, extensive libraries, and strong community support, making it a top choice for beginners and experienced developers alike.

Q2 :- What is an interpreter in Python?

Ans :- The Python interpreter is a program that executes your Python code line by line, translating it into machine-readable instructions and immediately running them. It's essential for making your Python code work and provides interactivity, error handling, and dynamic typing.

Q3 :- What are pre-defined keywords in Python?

Ans :- Pre-defined keywords in Python are reserved words with special meanings that cannot be used as identifiers. They are fundamental to Python's syntax and are used for tasks like controlling program flow, defining data structures, and handling exceptions. You can find a list of all keywords using import keyword; print(keyword.kwlist).

Q4 :- Can keywords be used as variable names?

Ans :- No, keywords cannot be used as variable names in Python because they have predefined meanings and purposes within the language. Using them as variable names would cause syntax errors and conflicts with Python's functionality.

Q5 :- What is mutability in Python?
Ans :- In Python, mutability refers to the ability of an object to be changed after it is created. If an object can be modified, it is considered mutable. If it cannot be changed without creating a new object, it is considered immutable.

Q6 :- Why are lists mutable, but tuples are immutable?

Ans :- Lists are mutable because they are designed for dynamic data manipulation, while tuples are immutable to ensure data integrity, hashability, and potential performance benefits.

Q7 :- What is the difference between “==” and “is” operators in Python?

Ans :- "==" checks for equality of values, while "is" checks for object identity (if they are the same object in memory). Use "==" for comparing values and "is" for checking if two variables refer to the same
object.

Q8 :- What are logical operators in Python?

Ans :- Logical operators (and, or, not) combine or modify Boolean expressions to control the flow of your programs based on multiple conditions.

Q9 What is type casting in Python?

Ans:- Type casting in Python is the process of converting a value from one data type to another, allowing you to perform operations or use data in ways specific to different types.

Q10 :- What is the difference between implicit and explicit type casting?

Ans:- Implicit type casting is done automatically by Python, while explicit type casting requires the use of built-in functions like int(), float(), str(), etc.

Q11 :- What is the purpose of conditional statements in Python?

ANS:- Conditional statements in Python determine the flow of code execution based on certain conditions. They allow your program to make decisions and execute different blocks of code depending on whether a condition is true or false.

Q12:- How does the elif statement work?

Ans:- he elif statement, short for "else if," is used in conjunction with if and else to define a series of conditional checks. If the initial if condition is False, Python checks the next elif condition and so on. If any elif condition is True, its code block is executed, and the remaining conditions are skipped. If none of the if or elif conditions are True, the else block (if present) is executed.In short, it allows for a sequence of "if this, then that, otherwise if this other thing, then that other thing" checks.

Q13:- What is the difference between for and while loops?

Ans :- for loops are used for iterating over a sequence, while while loops are used for repeating a block of code as long as a condition is True.
Essentially:
For: definite iteration (you know how many times you'll loop).
while: indefinite iteration (you loop until a condition changes).

Q14 :-Describe a scenario where a while loop is more suitable than a for loop.

Ans:- When you need to repeat a block of code until a specific condition becomes False, and you don't know beforehand how many iterations will be needed, a while loop is more appropriate.

For example, you might use a while loop to keep asking for user input until they enter a valid value or to read data from a file until the end of the file is reached. In such cases, you don't know the number of iterations in advance, so a while loop is a better choice than a for loop.

In essence, use while when the number of iterations is uncertain.

```python
#5 Create a list and tuple in Python, and demonstrate how attempting to change an element works differently
for each

# Creating a List
my_list = [1, 2, 3, 4, 5]

# Creating a Tuple
my_tuple = (1, 2, 3, 4, 5)

# Attempt to change an element in the List
my_list[0] = 10
print("List after modification:", my_list)

# Attempt to change an element in the Tuple (this will raise an error)
try:
    my_tuple[0] = 10
except TypeError as e:
    print("Error while modifying Tuple:", e)
```

```python
# Q14 Write a program to reverse a string using a while loop?

s = input("Enter a string: ")
reversed_string = ""
i = len(s) - 1

while i >= 0:
    reversed_string += s[i]
    i -= 1

print(reversed_string)
```

```python
# Q14 Write a Python program to calculate the factorial of a number provided by
# the user using a while loop?

n = int(input("Enter a number: "))
factorial = 1
i = 1

while i <= n:
    factorial *= i
    i += 1

print(factorial)
```
